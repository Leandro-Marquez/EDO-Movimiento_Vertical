## Parte B (Bifurcación)

## Puntos de Equilibrio
    Los puntos de equilibrio son los valores de v donde la velocidad deja de cambiar,
    es decir dv/dt = 0


## Resolucion de la Ecuacion

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores del parámetro µ
mu = np.linspace(-2, 4, 400)

# Puntos de equilibrio: µ ≥ 0
v_pos = np.sqrt(np.maximum(mu, 0))   # v = +√µ
v_neg = -np.sqrt(np.maximum(mu, 0))  # v = -√µ
```

## Construción del Gráfico de Bifurcación

```python
plt.figure(figsize=(8,6))
plt.plot(mu, v_pos, 'b', label='v = +√µ (equilibrio)')
plt.plot(mu, v_neg, 'r', label='v = -√µ (equilibrio)')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel('µ (parámetro)')
plt.ylabel('v (punto de equilibrio)')
plt.title('Puntos de equilibrio de dv/dt = µ - v²')
plt.legend()
plt.grid(True)
plt.show()
```
